```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 8)

# Normal distributed x and y vector with mean 0 and standard deviation 1
x = np.random.normal(0, 1, 500)
y = np.random.normal(0, 1, 500)
X = np.vstack((x, y)).T

plt.scatter(X[:, 0], X[:, 1])
plt.title('Generated Data')
plt.axis('equal');
```

```python
# Covariance
def cov(x, y):
    xbar, ybar = x.mean(), y.mean()
    return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)

# Covariance matrix
def cov_mat(X):
    return np.array([[cov(X[0], X[0]), cov(X[0], X[1])], \
                     [cov(X[1], X[0]), cov(X[1], X[1])]])

# Calculate covariance matrix 
cov_mat(X.T) # (or with np.cov(X.T))
```

```python
# Center the matrix at the origin
X = X - np.mean(X, 0)

# Scaling matrix
sx, sy = 0.7, 3.4
Scale = np.array([[sx, 0], [0, sy]])

# Apply scaling matrix to X
Y = X.dot(Scale)

plt.scatter(Y[:, 0], Y[:, 1])
plt.title('Transformed Data')
plt.axis('equal')

# Calculate covariance matrix
cov_mat(Y.T)
```

```python
# Scaling matrix
sx, sy = 0.7, 3.4
Scale = np.array([[sx, 0], [0, sy]])

# Rotation matrix
theta = 0.77*np.pi
c, s = np.cos(theta), np.sin(theta)
Rot = np.array([[c, -s], [s, c]])

# Transformation matrix
T = Scale.dot(Rot)

# Apply transformation matrix to X
Y = X.dot(T)

plt.scatter(Y[:, 0], Y[:, 1])
plt.title('Transformed Data')
plt.axis('equal');

# Calculate covariance matrix
cov_mat(Y.T)
```

```python
C = cov_mat(Y.T)
eVe, eVa = np.linalg.eig(C)

plt.scatter(Y[:, 0], Y[:, 1])
for e, v in zip(eVe, eVa.T):
    plt.plot([0, 3*np.sqrt(e)*v[0]], [0, 3*np.sqrt(e)*v[1]], 'k-', lw=2)
plt.title('Transformed Data')
plt.axis('equal');
```

```python
C = cov_mat(Y.T)

# Calculate eigenvalues
eVa, eVe = np.linalg.eig(C)

# Calculate transformation matrix from eigen decomposition
R, S = eVe, np.diag(np.sqrt(eVa))
T = R.dot(S).T

# Transform data with inverse transformation matrix T^-1
Z = Y.dot(np.linalg.inv(T))

plt.scatter(Z[:, 0], Z[:, 1])
plt.title('Uncorrelated Data')
plt.axis('equal');

# Covariance matrix of the uncorrelated data
cov_mat(Z.T)
```

```python
import random
import math

def euclidean_distance(a, b):
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(a, b)))

def initialize_centroids(data, k):
    return random.sample(data, k)

def assign_clusters(data, centroids):
    clusters = [[] for _ in centroids]
    for point in data:
        distances = [euclidean_distance(point, centroid) for centroid in centroids]
        closest_index = distances.index(min(distances))
        clusters[closest_index].append(point)
    return clusters

def update_centroids(clusters):
    new_centroids = []
    for cluster in clusters:
        if cluster:
            new_centroids.append([sum(dim)/len(cluster) for dim in zip(*cluster)])
        else:
            # If cluster is empty, randomly re-initialize a centroid
            new_centroids.append(random.choice(cluster))
    return new_centroids

def kmeans(data, k, max_iters=100):
    centroids = initialize_centroids(data, k)
    for _ in range(max_iters):
        clusters = assign_clusters(data, centroids)
        new_centroids = update_centroids(clusters)
        
        # Check convergence
        if all(euclidean_distance(c, nc) < 1e-4 for c, nc in zip(centroids, new_centroids)):
            break
        centroids = new_centroids
    return centroids, clusters

# Example usage:
data = [[1, 2], [2, 3], [3, 4], [10, 11], [11, 12], [12, 13]]
k = 2
centroids, clusters = kmeans(data, k)
print("Centroids:", centroids)
print("Clusters:", clusters)
```

```python
import math
import random

def gaussian_pdf(x, mean, cov):
    # Assuming diagonal covariance matrix: cov is a vector of variances
    dim = len(x)
    det_cov = 1
    for v in cov:
        det_cov *= v
    norm_const = 1 / math.sqrt((2 * math.pi) ** dim * det_cov)
    exp_val = 0
    for i in range(dim):
        exp_val += ((x[i] - mean[i]) ** 2) / cov[i]
    return norm_const * math.exp(-0.5 * exp_val)

def initialize_gmm(data, k):
    means = random.sample(data, k)
    covariances = []
    dim = len(data[0])
    for _ in range(k):
        # Initialize diagonal covariance with variance of each feature over all data
        variances = [sum((x[i] - sum(d[i] for d in data)/len(data))**2 for x in data)/len(data) for i in range(dim)]
        covariances.append(variances)
    weights = [1.0/k] * k
    return means, covariances, weights

def e_step(data, means, covariances, weights, k):
    responsibilities = []
    for x in data:
        resp = []
        total = 0
        for j in range(k):
            p = weights[j] * gaussian_pdf(x, means[j], covariances[j])
            resp.append(p)
            total += p
        resp = [r / total for r in resp]
        responsibilities.append(resp)
    return responsibilities

def m_step(data, responsibilities, k):
    n = len(data)
    dim = len(data[0])
    means = []
    covariances = []
    weights = []
    for j in range(k):
        resp_j = [r[j] for r in responsibilities]
        total_resp = sum(resp_j)
        
        # Update mean
        mean_j = [0] * dim
        for i in range(n):
            for d in range(dim):
                mean_j[d] += resp_j[i] * data[i][d]
        mean_j = [m / total_resp for m in mean_j]
        means.append(mean_j)
        
        # Update covariance (diagonal)
        cov_j = [0] * dim
        for i in range(n):
            for d in range(dim):
                cov_j[d] += resp_j[i] * ((data[i][d] - mean_j[d]) ** 2)
        cov_j = [c / total_resp for c in cov_j]
        covariances.append(cov_j)
        
        # Update weights
        weights.append(total_resp / n)
    return means, covariances, weights

def log_likelihood(data, means, covariances, weights, k):
    ll = 0
    for x in data:
        s = 0
        for j in range(k):
            s += weights[j] * gaussian_pdf(x, means[j], covariances[j])
        ll += math.log(s)
    return ll

def gmm(data, k, max_iters=100, tol=1e-4):
    means, covariances, weights = initialize_gmm(data, k)
    prev_ll = None
    for _ in range(max_iters):
        responsibilities = e_step(data, means, covariances, weights, k)
        means, covariances, weights = m_step(data, responsibilities, k)
        ll = log_likelihood(data, means, covariances, weights, k)
        if prev_ll is not None and abs(ll - prev_ll) < tol:
            break
        prev_ll = ll
    return means, covariances, weights

# Example usage:
data = [[1, 2], [2, 1], [3, 4], [8, 9], [9, 10], [10, 11]]
k = 2
means, covariances, weights = gmm(data, k)
print("Means:", means)
print("Covariances:", covariances)
print("Weights:", weights)
```

```python
import matplotlib.pyplot as plt

def plot_clusters(data, clusters, centroids):
    colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']  # Some colors for up to 7 clusters
    
    for idx, cluster in enumerate(clusters):
        cluster_points = list(zip(*cluster))  # unzip to separate x and y
        if cluster_points:
            plt.scatter(cluster_points[0], cluster_points[1], c=colors[idx % len(colors)], label=f'Cluster {idx+1}')
    
    centroid_points = list(zip(*centroids))
    plt.scatter(centroid_points[0], centroid_points[1], c='black', marker='x', s=100, label='Centroids')
    
    plt.legend()
    plt.show()

# Assuming data, clusters, centroids from K-Means:
plot_clusters(data, clusters, centroids)
```
